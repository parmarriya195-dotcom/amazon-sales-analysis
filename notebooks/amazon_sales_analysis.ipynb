**Convert Date Column** - Converts the order_date column to datetime format for proper date-based analysis.

```python
df["order_date"] = pd.to_datetime(df["order_date"])
```

**Extract Year and Month** - Creates new year and month columns from the order_date for time-based analysis.

```python
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
```

**Top 10 Products Revenue Chart** – Displays a bar chart of the top 10 products with the highest total revenue.


```python
top_products = df.groupby("product_id")["total_revenue"].sum().sort_values(ascending=False).head(10)

colors = plt.cm.Pastel2.colors 

top_products.plot(
    kind="bar",
    figsize=(10,8),
    color=colors
)

plt.title("Top 10 Products by Revenue", fontsize=16)
plt.xlabel("Product_ID", fontsize=14)
plt.ylabel("Total_Revenue", fontsize=14)

plt.show()
```

```python
product = df.groupby("product_category")["profit"].sum().sort_values(ascending=False)

color = plt.cm.Pastel1.colors

product.plot(
    kind="bar",
    figsize=(9,8),
    color=colors
)

plt.title("Profit by Product_category", fontsize=16)
plt.xlabel("Product_Category", fontsize=14)
plt.ylabel("Profit", fontsize=14)

plt.tight_layout()
plt.show()
```

**Category Revenue Pie Chart** – Displays a pie chart showing the percentage share of total revenue by each product category.


```python
category = df.groupby("product_category")["total_revenue"].sum()

category.plot(
    kind="pie",
    autopct = "%1.1f%%",
    figsize=(8,8,),
    colors=plt.cm.Pastel1.colors, 
    startangle = 90
)

plt.title("Revenue by Category", fontsize=16)
plt.ylabel("")
plt.tight_layout()
plt.show()
```

**Quantity Sold Over Time** – Plots a line chart showing how total quantity sold changes over time based on order date.


```python
plt.figure(figsize=(10,8))

data = df.groupby("order_date")["quantity_sold"].sum()

data.plot(
    marker='o',
    linestyle='-'
)

plt.title("Quantity Sold Over Time", fontsize=16)
plt.xlabel("Order Date", fontsize=14)
plt.ylabel("Quantity Sold", fontsize=14)

plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

**Quantity Sold by Category Chart** – Displays a bar chart of total quantity sold for each product category in descending order.


```python
quantity = df.groupby("product_category")["quantity_sold"].sum().sort_values(ascending=False)

quantity.plot(
    kind="bar",
    figsize=(10,8),
    color=plt.cm.Dark2.colors
)

plt.title("Quantity_Sold by Category", fontsize=16)
plt.xlabel("Product_Category", fontsize=14)
plt.ylabel("Quantity_sold", fontsize=14)
plt.tight_layout()
plt.show()
```

**Discount vs Quantity Sold Scatter Plot** – Shows the relationship between discount percentage and quantity sold using a scatter plot.


```python
plt.figure(figsize=(10,8))

plt.scatter(
    df["discount_percent"],
    df["quantity_sold"],
    alpha=0.6
)

plt.xlabel("Discount Percentage", fontsize=14)
plt.ylabel("Quantity Sold", fontsize=14)
plt.title("Discount vs Quantity Sold", fontsize=16)

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

**Quantity Sold by Region Pie Chart** – Displays a pie chart showing the percentage distribution of quantity sold across different regions.


```python
region = df.groupby("customer_region")["quantity_sold"].sum()

region.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(8,8),
    colors=plt.cm.Pastel1.colors,
    startangle = 90
)

plt.title("Quantity Sold by Region", fontsize=16)
plt.ylabel("")
plt.tight_layout()
plt.show()
```

**Revenue Over Time Line Chart** – Plots a line chart showing how total revenue changes over time based on order date.

**Price Distribution Histogram** – Displays a histogram showing the distribution of product prices across different ranges.


```python
plt.figure(figsize=(10,6))

plt.hist(df["price"], bins=10)

plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

**Detailed Correlation Heatmap** – Displays a heatmap with annotated correlation values between numerical features using a color gradient.


```python
plt.figure(figsize=(10,6))

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.show()
```
